# init.py 
Initialies the program with fake student names and classes from UMD by using different APIs. 

#### Fake names are generated using fakename API available over the internet.   

#### UMD courses are obtained by the UMD open-source API developed and maintained by some of my friends over at University of Maryland. The APIs make it easier to obtain the schedule of classes data which is scattered all around the UMD webpages, without manually scraping it from the official website "testudo". 

```python
# All imports
import numpy as np
import requests
import time
import sqlite3
```

```python
start = time.time()
ids = np.random.randint(1000000000, 2000000000, 500)
names = []
for i in range(500):
    strn = "https://api.namefake.com/english-united-states/" + ("female/" if np.random.randint(0,10) < 5 else "male/")
    x = requests.get(strn)
    names.append(x.json()["name"])

end = time.time()
```

```python
print('Time taken: ' + str(end - start) + " seconds, or: " +  str( (end - start) / 60 ) + " minutes." )
```

```python
names[69] = "Tejan Gandhi"
```

```python
with open('students.csv', 'w') as f:
    for i in range(500):
        f.write(str(ids[i]) + "," + names[i] + "\n")
f.close()
```

```python
courses = ["CMSC330", "CMSC351", "CMSC420", "CMSC451", "CMSC216",
           "CMSC132", "MATH140", "MATH240", "PHYS161", "CHEM131",
          "CMSC335", "CMSC320", "CMSC470", "STAT400"]
start = time.time()
meetings = []
for course in courses:
    info = requests.get("https://api.umd.io/v1/courses/" + course).json()
    if len(info) > 0 and len(info[0]["sections"]) > 0:
        section = info[0]["sections"][0]
    meeting_times = requests.get("https://api.umd.io/v1/courses/" + course + "/sections/" +
                                 section[8:]).json()[0]["meetings"]
    
    meetings.append(meeting_times)
end = time.time()
```

```python
time_taken = end - start
print("Time taken: " + str(time_taken) + " and minutes: " + str(time_taken / 60))
with open('classes.csv', 'w') as f2:
    for i in range(len(courses)):
        str_temp = courses[i] + ": ["
        # f2.write(course + ": [")
        for j in range(len(meetings[i])):
            str_temp += ( meetings[i][j]["days"] + "|" + meetings[i][j]["building"]+meetings[i][j]["room"] + "|" + meetings[i][j]["start_time"] + "|" + meetings[i][j]["end_time"] +", ")
        str_temp = str_temp[:-2]
        str_temp += "]"
        print(str_temp)
        f2.write(str_temp + "\n")
f2.close()
```

## Storing the data into a SQLite database
Since there are 500 students, it is wise to store the students data into a SQL database  
for faster queries.   
  
It is fine for courses data since we have less datapoints. It'll be faster to read and store courses data,   
on RAM in the program.  

```python
ids_arr = [int(i) for i in ids]
c = sqlite3.connect('students_db') 
cursor = c.cursor()
cursor.execute('''
          CREATE TABLE IF NOT EXISTS students
          ([Student_ID] INTEGER PRIMARY KEY, [Student_Name] TEXT)
          ''')
cursor.executemany("""
    INSERT INTO students ('Student_ID', 'Student_Name')
    VALUES (?, ?)""", list(zip(ids_arr,names)))
```

```python
cursor.execute("SELECT * FROM students") 
cursor.fetchall()
```

```python
c.commit()
c.close()
```
